# Major Oxide Boxplots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## By Sample Type

```python
#import data

df = pd.read_csv(r'C:\Users\ctayl\Desktop\Malartic\Python\Malartic_geochem_majoroxides_redone.csv')

df.head()
```

```python
fig, axis=plt.subplots(3,3, figsize=(16,14))

#Add space between rows of plots
plt.subplots_adjust(hspace = 0.2)

#plot 1
ax=sns.boxplot(x='Sample', y='Al2O3', data=df, ax=axis[0,0], palette="Blues",
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='$\mathregular{Al_2O_3}$ (wt%)')

#plot 2
ax=sns.boxplot(x='Sample', y='CaO', data=df, ax=axis[0,1], palette="Blues", 
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='CaO (wt%)')

#plot 3
ax=sns.boxplot(x='Sample', y='Fe2O3', data=df, ax=axis[0,2], palette="Blues", 
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='$\mathregular{Fe_2O_3}$ (wt%)')

#plot 4
ax=sns.boxplot(x='Sample', y='K2O', data=df, ax=axis[1,0], palette="Blues",
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='$\mathregular{K_2O}$ (wt%)')

#plot 5
ax=sns.boxplot(x='Sample', y='MgO', data=df, ax=axis[1,1], palette="Blues",
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='MgO (wt%)')

#plot 6
ax=sns.boxplot(x='Sample', y='MnO', data=df, ax=axis[1,2], palette="Blues", 
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='MnO (wt%)')

#plot 7
ax=sns.boxplot(x='Sample', y='Na2O', data=df, ax=axis[2,0], palette="Blues", 
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='$\mathregular{Na_2O}$ (wt%)')

#plot 8
ax=sns.boxplot(x='Sample', y='P2O5', data=df, ax=axis[2,1], palette="Blues", 
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='$\mathregular{P_2O_5}$ (wt%)')

#plot 9
ax=sns.boxplot(x='Sample', y='TiO2', data=df, ax=axis[2,2], palette="Blues", 
               order=["Surface (2016)", "Surface (2015)", 
                      "Pionjar (2015)"]).set(ylabel='$\mathregular{TiO_2}$ (wt%)')

#plt.savefig('major_oxide_sample_type', transparent=False, bbox_inches='tight')
```

## By Degree of Oxidation

```python
fig, axis=plt.subplots(3,3, figsize=(16,14))

#Add space between rows of plots
plt.subplots_adjust(hspace = 0.2)


#plot 1
ax=sns.boxplot(x='Oxidation', y='Al2O3', data=df, ax=axis[0,0], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='$\mathregular{Al_2O_3}$ (wt%)')

#plot 2
ax=sns.boxplot(x='Oxidation', y='CaO', data=df, ax=axis[0,1], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='CaO (wt%)')

#plot 3
ax=sns.boxplot(x='Oxidation', y='Fe2O3', data=df, ax=axis[0,2], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='$\mathregular{Fe_2O_3}$ (wt%)')

#plot 4
ax=sns.boxplot(x='Oxidation', y='K2O', data=df, ax=axis[1,0], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='$\mathregular{K_2O}$ (wt%)')

#plot 5
ax=sns.boxplot(x='Oxidation', y='MgO', data=df, ax=axis[1,1], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='MgO (wt%)')

#plot 6
ax=sns.boxplot(x='Oxidation', y='MnO', data=df, ax=axis[1,2], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='MnO (wt%)')

#plot 7
ax=sns.boxplot(x='Oxidation', y='Na2O', data=df, ax=axis[2,0], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='$\mathregular{Na_2O}$ (wt%)')

#plot 8
ax=sns.boxplot(x='Oxidation', y='P2O5', data=df, ax=axis[2,1], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='$\mathregular{P_2O_5}$ (wt%)')

#plot 9
ax=sns.boxplot(x='Oxidation', y='TiO2', data=df, ax=axis[2,2], palette="Reds", 
               order=["Unapparent", "Slight", "High"]).set(ylabel='$\mathregular{TiO_2}$ (wt%)')


#plt.savefig('major_oxide_oxidation', transparent=False, bbox_inches='tight')
```
