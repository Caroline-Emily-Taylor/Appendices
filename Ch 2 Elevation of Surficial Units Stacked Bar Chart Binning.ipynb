## Binning Elevation Values

```python
# https://towardsdatascience.com/data-preprocessing-with-python-pandas-part-5-binning-c5bd5fd1b950
#Find all min and max values

#Make a df of all minimum values
print("Minimum: ")
print(df.min())


#whitespace
print(" ")


#Make a df ol all maximum values
print("Maximum: ")
print(df.max())
```

```python
#Find min and max values for binning


#Make a df of all minimum values
df_min = df.min()

print("Minimum: ")
min_value = df_min.min()
print(min_value)


#whitespace
print(" ")


#Make a df of all maximum values
df_max = df.max()

print("Maximum: ")
max_value = df_max.max()
print(max_value)
```

```python
# Use min and max to set n number of edges (i.e., # of bins +1) based on distance

#For 35 bins use 36 edges
bins = np.linspace(min_value, max_value, 36)

bins
```

## Bin Counts Based on Elevation Data by Sediment Type

```python
#Create empty dataframe for bin values

df_bins = pd.DataFrame()
```

```python
#Bin df based on elevation
#"Glaciolacustrine", "Till", "Bedrock", "Organic", "Alluvial", "Esker"
#"Deep water", "Deep-to-shallow", "Shallow water", "Continuous", "Discontinuous"

#'for loop' here?

df_bins['Till'] = pd.cut(x = df['Till'], bins = bins, include_lowest = True)
df_bins['Glaciolacustrine'] = pd.cut(x = df['Glaciolacustrine'], bins = bins, include_lowest = True)
df_bins['Bedrock'] = pd.cut(x = df['Bedrock'], bins = bins, include_lowest = True)
df_bins['Organic'] = pd.cut(x = df['Organic'], bins = bins, include_lowest = True)
df_bins['Alluvial'] = pd.cut(x = df['Alluvial'], bins = bins, include_lowest = True)
df_bins['Esker'] = pd.cut(x = df['Esker'], bins = bins, include_lowest = True)
df_bins['Deep water'] = pd.cut(x = df['Deep water'], bins = bins, include_lowest = True)
df_bins['Deep-to-shallow'] = pd.cut(x = df['Deep-to-shallow'], bins = bins, include_lowest = True)
df_bins['Shallow water'] = pd.cut(x = df['Shallow water'], bins = bins, include_lowest = True)
df_bins['Continuous'] = pd.cut(x = df['Continuous'], bins = bins, include_lowest = True)
df_bins['Discontinuous'] = pd.cut(x = df['Discontinuous'], bins = bins, include_lowest = True)
```
