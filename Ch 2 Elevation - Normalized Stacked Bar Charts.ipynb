# Stacked Bar Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Stacked Bar Chart of Main Units

```python
#Import data

main = pd.read_csv(r'C:\Users\ctayl\Desktop\Malartic\Python\elevation_main_norm.csv')

main
```

```python
#Stacked bar chart, main units

colors = ['#e8718d', '#8cac7e', '#987ea5', '#e1ab00', '#bebebe', '#e5d11e']


main.plot(x = 'Elevation',
          kind='bar', 
          stacked=True, 
          color = colors,
          figsize = (9,7))


#format legend
plt.legend(loc='upper center',
           bbox_to_anchor=(0.5, -0.12), 
           ncol=6)


plt.xlabel('Elevation (masl)', fontsize=12)

plt.ylabel('Percent', fontsize=12)

plt.show()
```

## Stacked Bar Chart with Subunits

```python
#Import data

subs = pd.read_csv(r'C:\Users\ctayl\Desktop\Malartic\Python\elevation_subs_norm.csv')

subs
```

```python
#Stacked bar chart, including subunits

colors = ['#e8718d', '#33a02c', '#b2df8a', '#d7ade1', '#be82cd', '#9f62ae', '#e1ab00', 'silver', 'gold']

subs.plot(x='Elevation',
          kind='bar', 
          stacked=True, 
          color=colors,
          figsize=(9,7))


#format legend
plt.legend(loc='upper center',
           bbox_to_anchor=(0.5, -0.2), 
           ncol=3)


plt.xlabel('Elevation (masl)', fontsize=12)
plt.ylabel('Percent', fontsize=12)

#plt.savefig('elevation_seds', transparent=False, bbox_inches='tight')

plt.show()
```

```python
#Import frequency data

freq = pd.read_csv(r'C:\Users\ctayl\Desktop\Malartic\Python\elevation_subs.csv')

freq.head()
```

```python
#Add a second y-axis to show relative frequency
#Stacked bar chart, including subunits

colors = ['#e8718d', '#33a02c', '#b2df8a', '#d7ade1', '#be82cd', '#9f62ae', '#e1ab00', 'silver', 'gold']

ax = subs.plot(x='Elevation',
          kind='bar', 
          stacked=True, 
          color=colors,
          figsize=(8,5.5))

plt.ylabel('Abundace by Unit Type (%)', fontsize=12)
plt.xlabel('Elevation (masl)', fontsize=12)

#second axis
#Keep getting error: No handles with labels found to put in legend.


# twin object for two different y-axis on the sample plot
ax2=ax.twinx()

# make a plot with different y-axis using second axis object
ax2.plot(freq.Elevation, freq["Frequency"], color="black")
plt.ylabel('Relative Frequency of Area (%)', fontsize=12)




#format legend
ax.legend(loc='upper center',
           bbox_to_anchor=(0.5, -0.25), 
           ncol=3,
           columnspacing=3.4)





#plt.savefig('elevation_seds', transparent=False, bbox_inches='tight')

plt.show()
```
