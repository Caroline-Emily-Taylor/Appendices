```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Local ice flow indicators

```python
#import color palettes
import seaborn as sns

pal = sns.color_palette("Blues", 4)
print("Blues: ", pal.as_hex())
```

## Small-scale

```python
#Import a csv file
df = pd.read_csv(r'C:\Users\ctayl\Desktop\Malartic\Python\Malartic_local_ice_flow_smallscale_all.csv')
df.head()
```

```python
#Create an array from the 'trend' column of the dataframe and display horizontally (one line)
Trend=df.loc[:,'Trend']
Trend=np.hstack(Trend)
print(Trend)
```

```python
#First, let's prepare the data to build the rose diagram
#Calculate the number of directions (trends) every 10° using numpy.histogram.
#numpy.arange(start, stop, step)
bin_edges = np.arange(-5, 366, 10)                              
number_of_trends, bin_edges = np.histogram(Trend, bin_edges)

#37 values
print("Trends every 10°:")
print(number_of_trends)

#Whitespace
print("\n")

print("Bin edges:")
print(bin_edges)
```

```python
#36 values
Trends=number_of_trends[0:-1]
print(Trends)                    
```

```python
#Plot the data
fig = plt.figure(figsize=(14,14))

ax = fig.add_subplot(222, projection='polar')

ax.bar(np.deg2rad(np.arange(0, 360, 10)), Trends, 
       width=np.deg2rad(10), bottom=0.0, color='#d0e1f2', edgecolor='k', label="Small-scale")
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)
ax.set_thetagrids(np.arange(0, 360, 10), labels=np.arange(0, 360, 10))
ax.set_rgrids(np.arange(1, Trends.max() + 1, 2), angle=0, weight= 'black')

#Title size
#ax.set_title('Small-scale Ice Flow Indicators', size=20, pad=20)

#ax.legend(loc="upper right", bbox_to_anchor=(1.1, 0))
```
